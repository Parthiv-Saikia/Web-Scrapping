```python
import requests
import json
import csv
from bs4 import BeautifulSoup

base_url = "https://www.moneycontrol.com/news/tags/tata-steel.html"
headers = {
    "User-Agent": "Mozilla/5.0"
}

page = 1
news_items = []

while True:
    if page == 1:
        url = base_url
    else:
        url = f"{base_url}/page-{page}/"

    print(f"Fetching: {url}")
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    found_items = 0
    for script in soup.find_all('script', type='application/ld+json'):
        try:
            data = json.loads(script.string)
            if data.get('@type') == 'ItemList':
                for item in data.get('itemListElement', []):
                    url = item.get('url')
                    name = item.get('name')
                    if url and name:
                        news_items.append({'URL': url, 'Title': name})
                        found_items += 1
                break  # Stop after processing ItemList
        except:
            continue

    if found_items == 0:
        print("No more items found. Stopping.")
        break

    page += 1

# Write to CSV
with open('tata_steel_all_news_Javascript.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['URL', 'Title'])
    writer.writeheader()
    writer.writerows(news_items)

print(f"Total news items extracted: {len(news_items)}")
print("Saved to tata_steel_all_news.csv")
```
